 - données : f continue sur un intervalle ,a et b tels que f(a)f(b) <= 0
 
 - résultat : une approximation à epsilon près d'une solution de l'équation         f(x)=0 obtenue par dichotomie 

```python

def dicho(a,b,f,n):
 
    g , d = min (a , b) , max (a , b)
    while d - g > n :
            m = ( g + d )/2
            if f( m ) == 0:
                return m
            elif f( g ) * f( m ) < 0:
                d = m
            else :
                g = m
    return ( g + d )/2

```

### fonction dichow calculer iteration 

```python
def dichow(a,b,f,eps):
    """version tant que qui renvoila solution avec le nombre d'''itération """
    
    g , d = min (a , b) , max (a , b)
    nb=0
    while d - g > eps :
            nb+=1
            m = ( g + d )/2
            if f( m ) == 0:
                return m
            elif f( g ) * f( m ) < 0:
                d = m
            else :
                 g = m
    print (" nombre d'ité rations né cessaires :", nb )       
       
        
    return ( g + d )/2

    
```

## code python point fixe

```python
from sympy import *
import numpy as np
f=lambda x:  x**3+4*x**2-10 # la fonction f
g1=lambda x: 1/2*sqrt(10-x**3)
g2=lambda x: sqrt(10/(x+4))
g3=lambda x: x-x**3-4*x**2 + 10


def ptfixe(g,x0,eps,n):
    
    n=0
    x1=g(x0)
    while abs(x1-x0)>eps:
        n+=1
        x0=x1
        x1=g(x0)
    print("nombre d'iteration", n)
    return x1



print('pas de convergence avec le nombre d''iteration indique')
```

```python

x1= ptfixe(g1,1.5,0.001,30)
x2= ptfixe(g2,1.5,0.001,30)
x3= ptfixe(g3,1.5,20,30)
print(x1)
print(x2)
print(x3)
```

### EXERCICE 4:

- Données : f continue sur un intervalle , a : valeur approchée d'un zéro de f
-  Résultat : une approximation d'une solution de l'é quation f(x)=0 obtenue par la mé thode de Newton .
- Test d'arrêt : deux termes consécutifs x,y tels que |x-y| <= epsilon     Affiche le nombre d'ité rations effectuées.

```python
def newton (f , df ,a , epsilon ):

    x = a
    y = x - f ( x )/ df ( x )
    n = 0
    while abs(y - x ) > epsilon and n < 50:
        n += 1
        z = y - f ( y ) / df ( y )
        x , y = y , z
    if n < 50:
        print (" nombre d'ité rations né cessaires :", n )
        return y
    else :
        print (" aucune solution trouv ée en 50 ité rations ")
```

```python
import math
f=lambda x:math.exp(-x)-x
df=lambda x:-math.exp(-x)-1

```

```python
sol1=newton(f,df,0,0.001)
sol2=newton(f,df,1,0.001)
print(sol1)
print(sol2)
```
